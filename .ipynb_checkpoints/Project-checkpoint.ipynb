```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('c:/Users/HP/Desktop/Crime_Data.csv')

# Convert the 'DATE OCC' column to datetime
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')

# Extract the year from the 'DATE OCC' column
df['YEAR'] = df['DATE OCC'].dt.year

# Count the number of crimes per year
crime_counts = df['YEAR'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(crime_counts, labels=crime_counts.index, autopct='%1.1f%%')
plt.title('Number of Crimes by Year')
plt.axis('equal')
plt.legend(title='Year', loc='upper left')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('c:/Users/HP/Desktop/Crime_Data.csv')

# Convert the 'DATE OCC' column to datetime
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')

# Extract the year from the 'DATE OCC' column
df['Year'] = df['DATE OCC'].dt.year

# Create age intervals with a 5-year gap
df['Vict Age'] = pd.cut(df['Vict Age'], bins=range(0, 101, 5), labels=[f"{i} _ {i+4}" for i in range(0, 100, 5)])

# Group the data by year and victim age, then count the number of crimes
crime_counts = df.groupby(['Year', 'Vict Age']).size().unstack(fill_value=0)

# Rename the columns to remove special characters
crime_counts.columns = crime_counts.columns.str.replace('-', '')

# Create the line graph
fig, ax = plt.subplots(figsize=(12, 6))
crime_counts.plot(kind='line', ax=ax)
ax.set_title('Number of Crimes by Year and Victim Age (Line Graph)')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Crimes')
ax.legend(title='Victim Age')
plt.show()
```

```python
df = pd.read_csv('c:/Users/HP/Desktop/Crime_Data.csv')
# Convert the 'DATE OCC' column to datetime
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')

# Extract the year from the 'DATE OCC' column
df['Year'] = df['DATE OCC'].dt.year

# Group the data by year and victim sex, then count the number of crimes
crime_counts = df.groupby(['Year', 'Vict Sex']).size().unstack(fill_value=0)

# Create a bar chart
fig, ax = plt.subplots(figsize=(12, 6))
crime_counts.plot(kind='bar', ax=ax)

# Set the chart title and axis labels
ax.set_title('Number of Crimes by Year and Victim Sex')
ax.set_xlabel('Victim Sex')
ax.set_ylabel('Number of Crimes')
ax.legend(title='Year')

# Show the chart
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('c:/Users/HP/Desktop/Crime_Data.csv')

# Convert the 'DATE OCC' column to datetime
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')

# Extract the day of week, month, and year from the 'DATE OCC' column
df['Day of Week'] = df['DATE OCC'].dt.day_name()
df['Month'] = df['DATE OCC'].dt.month
df['Year'] = df['DATE OCC'].dt.year

# Count the daily crime incidents
daily_counts = df.groupby('Day of Week').size().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Create the bar plot
plt.figure(figsize=(12, 8))  # Increase the figure size
colors = ['#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', '#46f0f0']
sns.barplot(x=daily_counts.index, y=daily_counts, palette=colors, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Daily Trends in Crime Incidents')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.tight_layout()  # Adjust the layout to fit the content
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.scrapethissite.com/pages/forms/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table element
table = soup.find('table')

# Create an empty list to store the data
data = []

# Iterate through the rows of the table
for row in table.find_all('tr'):
    # Extract the data from each cell
    cells = row.find_all('td')
    if len(cells) == 9:
        team_name = cells[0].text.strip()
        year = cells[1].text.strip()
        wins = cells[2].text.strip()
        losses = cells[3].text.strip()
        ot_losses = cells[4].text.strip()
        win_percentage = cells[5].text.strip()
        goals_for = cells[6].text.strip()
        goals_against = cells[7].text.strip()
        goal_differential = cells[8].text.strip()
        
        # Append the data to the list
        data.append([team_name, year, wins, losses, ot_losses, win_percentage, goals_for, goals_against, goal_differential])

# Create a pandas DataFrame from the data
df = pd.DataFrame(data, columns=['Team', 'Year', 'Wins', 'Losses', 'OT Losses', 'Win %', 'Goals For', 'Goals Against', 'Goal Differential'])

# Print the DataFrame
print(df)
```
